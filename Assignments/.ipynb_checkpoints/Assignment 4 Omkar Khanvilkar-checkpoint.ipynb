```python
feature_col = ['SepalLength', 'SepalWidth', 'PetalLength','PetalWidth']
X = df[feature_col].values
y = df['target'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =30)
```

```python
from sklearn.preprocessing import StandardScaler
# Load the standard scaler
sc = StandardScaler()
```

```python
# Compute the mean and standard deviation based on the training data
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
print('Accuracy of linear SVC on training set: {:.2f}'.format(clf.score(X_train, y_train)))
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
import numpy as np
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python

from sklearn.svm import SVC
svc = SVC(kernel="rbf")
svc.fit(X_train, y_train)
```

```python
accuracy_score(y_test, svc.predict(X_test))
```

```python

from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
```

```python
grid.fit(X_train, y_train)
accuracy_score(y_test, grid.predict(X_test))
```

```python
# Fitting clasifier to the Training set
# Loading libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score

# for the learning model we are taking the value of k as 5.
# I tried all the values from 2 to 10 most of them are giving the 100% accuracy which is nothing but model overfitting.
# When i am putting k as 5 it is giving me 97%~ accuracy.
classifier = KNeighborsClassifier(n_neighbors=9)

# Fitting the model
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)*100
```

```python
accuracy
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
# configure the cross-validation procedure
cv = KFold(n_splits=3, shuffle=True, random_state=1)

RandomizedSearchCV(classifier,cv, n_iter =10)
```
